```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt


def sufficient_sample_size(alpha, margin_of_error, std_dev):
    z_score = stats.norm.ppf(1 - (1 - alpha) / 2)
    return int(np.ceil((z_score * std_dev / margin_of_error) ** 2))

def visualize_results(distribution_name, sample, q_1, q_95, ci_lower, ci_upper):
    """Визуализирует выборку, квантили и доверительные интервалы."""
    plt.figure(figsize=(10, 6))
    plt.hist(sample, bins=30, alpha=0.7, label="Выборка", color="blue", edgecolor="black")
    plt.axvline(q_1, color="red", linestyle="--", label="1% квантиль")
    plt.axvline(q_95, color="green", linestyle="--", label="95% квантиль")
    plt.axvline(ci_lower, color="orange", linestyle=":", label="Нижняя граница ДИ")
    plt.axvline(ci_upper, color="purple", linestyle=":", label="Верхняя граница ДИ")
    plt.title(f"Распределение: {distribution_name}")
    plt.xlabel("Значения")
    plt.ylabel("Частота")
    plt.legend()
    plt.grid()
    plt.show()

alpha_values = [0.90, 0.95, 0.99]
delta_values = [0.1, 0.05, 0.01]
distributions = {
    'Равномерное': (np.random.uniform, (0, 1)),
    'Нормальное': (np.random.normal, (0, 1)),
    'Коши': (np.random.standard_cauchy, ())
}
std_dev_assumption = 1

for dist_name, (dist_func, dist_params) in distributions.items():
    print(f"\nРаспределение: {dist_name}")
    for delta in delta_values:
        for alpha in alpha_values:
            sample_size = sufficient_sample_size(alpha, delta, std_dev_assumption)

            sample = dist_func(*dist_params, size=sample_size)

            q_1 = np.percentile(sample, 1)
            q_95 = np.percentile(sample, 95)
            mean = np.mean(sample)
            std_err = np.std(sample, ddof=1) / np.sqrt(sample_size)
        
            critical_value = stats.norm.ppf(1 - (1 - alpha) / 2)
        
            ci_lower = mean - critical_value * std_err
            ci_upper = mean + critical_value * std_err
            
            within_bounds = ci_lower >= np.min(sample) and ci_upper <= np.max(sample)
            
            print(f"Alpha: {alpha}, Delta: {delta}, Размер выборки: {sample_size}")
            print(f"  1% квантиль: {q_1:.3f}, 95% квантиль: {q_95:.3f}")
            print(f"  Доверительный интервал: ({ci_lower:.3f}, {ci_upper:.3f}), "
                  f"Попадает в границы: {'Да' if within_bounds else 'Нет'}")
            
            # Визуализация
            visualize_results(dist_name, sample, q_1, q_95, ci_lower, ci_upper)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_samples(distribution_func, params, size):
    return distribution_func(*params, size=size)

def monte_carlo_uncertainty(samples, n_iterations=10000):
    bootstrap_means = [np.mean(np.random.choice(samples, size=len(samples), replace=True))
                       for _ in range(n_iterations)]
    return np.std(bootstrap_means)

def linearization_uncertainty(samples):
    mean = np.mean(samples)
    std_dev = np.std(samples, ddof=1)

    error_mean = std_dev / np.sqrt(len(samples))

    variance = np.var(samples, ddof=1)
    error_variance = np.sqrt(2 * variance**2 / (len(samples) - 1))

    return error_mean, error_variance

def kreinovich_uncertainty(samples):
    mean = np.mean(samples)
    std_dev = np.std(samples, ddof=1)

    error_mean = std_dev / np.sqrt(len(samples))

    variance = np.var(samples, ddof=1)
    error_variance = (2 * variance**2) / (len(samples) - 1)

    return error_mean, error_variance

distribution_func = np.random.normal
params = (0, 1) 
sample_size = 100
samples = generate_samples(distribution_func, params, sample_size)

monte_carlo_error = monte_carlo_uncertainty(samples)
linearization_errors = linearization_uncertainty(samples)
kreinovich_errors = kreinovich_uncertainty(samples)

print(f"Метод Монте-Карло: Погрешность = {monte_carlo_error:.4f}")
print(f"Метод линеаризации: Погрешность математического ожидания = {linearization_errors[0]:.4f}, "
      f"Погрешность дисперсии = {linearization_errors[1]:.4f}")
print(f"Метод Крейновича: Погрешность математического ожидания = {kreinovich_errors[0]:.4f}, "
      f"Погрешность дисперсии = {kreinovich_errors[1]:.4f}")

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(samples, bins=20, alpha=0.7, color='skyblue', edgecolor='black')
plt.axvline(np.mean(samples), color='red', linestyle='dashed', label='Среднее')
plt.title('Распределение выборки')
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.legend()

methods = ['Монте-Карло', 'Линеаризация (МО)', 'Линеаризация (Дисп.)',
           'Крейнович (МО)', 'Крейнович (Дисп.)']
errors = [monte_carlo_error, linearization_errors[0], linearization_errors[1],
          kreinovich_errors[0], kreinovich_errors[1]]

plt.subplot(1, 2, 2)
plt.bar(methods, errors, color='lightcoral', edgecolor='black')
plt.title('Погрешности, оцененные разными методами')
plt.ylabel('Величина погрешности')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
